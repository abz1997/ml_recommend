```python
import pandas as pd

df = pd.read_csv ('clean_data.csv',lineterminator='\n')
df
```

```python
# we have to encode the categorical data so we can use techniques such as one hot encoding, label encoding, target encoding.
# Though label encoding is straight but it has the disadvantage that the numeric values can be misinterpreted by algorithms 
# as having some sort of hierarchy/order in them. We will thus use one hot encoding however this will result in many new columns.

df['category2'].describe()
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')
enc_df = pd.DataFrame(enc.fit_transform(df[['category2']]).toarray())
new_df = df.join(enc_df)
new_df
```

```python
# using dummy values allows you to encode as many category columns as you would like and choose how to label the columns using a prefix
#, prefix=["category0_is", 'category1_is', 'category2_is', 'category3_is']
dum_df = pd.get_dummies(df, columns=['category0', 'category1', 'category2', 'category3'], prefix=["category0_is", 'category1_is', 'category2_is', 'category3_is'] )# merge with main df  on key values
dummies_df = df.merge(dum_df)
dummies_df
```

```python
dum2_df = pd.get_dummies(df, columns=['category0', 'category1'], prefix=["category0_is", 'category1_is'] )# merge with main df  on key values
cat1_2_df = df.merge(dum2_df)
cat1_2_df
```

```python
dum_df = pd.get_dummies(df, columns=['category0'], prefix=["category0_is"] )# merge with main df  on key values
cat0_df = df.merge(dum_df)
cat0_df
```

```python
dum_df = pd.get_dummies(df, columns=['category1'], prefix=["category1_is"] )# merge with main df  on key values
cat1_df = df.merge(dum_df)
cat1_df
```

```python
#lets test it with only dummy variables for category 0
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import metrics

features = ['price','id','product_name', 'product_description','location', 'page_id', 'create_time', 'category0', 'category1', 'category2', 'category3']

#X = data.drop('price',axis =1)
X = cat0_df.drop(columns = features)
y = cat0_df['price']

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3)

print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
print(test_y.shape)
```

```python
model = linear_model.LinearRegression()

model.fit(train_X, train_y)
y_pred = model.predict(test_X)
```

```python
from sklearn.metrics import mean_squared_error, r2_score 
import numpy as np

def calculate_metrics(y_test, y_pred):  
    ''' 
    y_test:- ground truth 
    y_pred:- model predict  
    '''
    mse = mean_squared_error(test_y, y_pred) 
    rmse = np.sqrt(mse) 
    r2_scores = r2_score(test_y, y_pred) 
    
    print("MSE: ", mse) 
    print("RMSE: ", rmse) 
    print("R2_score: ", r2_scores) 
    
calculate_metrics(test_y, y_pred)
```
