```python
min_width = 10000
min_height = 10000

for img in list_img[0:50]:
    size = Image.open(img).size
    width = size[0]
    height = size[1]
    if width < min_width:
        min_width = width
    if height < min_height:
        min_height = height

print(min_width, min_height)
```

```python
final_size = 512
small_size = 256
smaller_size = 28
#for n, img in enumerate(list_img[0:3],1):
# pastes saved images on black image at centre
for img, file in enumerate(list_img): 
    if file.endswith('.jpg'):
        next_img = Image.open(file)
        # fn, fext = os.path.splitext(img)
        # print(fn,fext)
        black_image = Image.new(mode='RGB', size=(smaller_size,smaller_size))
        prev_size = next_img.size #has height and width
        max_dim = max(next_img.size) #gets max value from height and width
        ratio = smaller_size / max_dim #finds ratio
        new_img_size = (int(prev_size[0]*ratio), (int(prev_size[1]*ratio))) 
        new_img = next_img.resize(new_img_size) #resizes the image
        black_image.paste(new_img, ((smaller_size - new_img_size[0])//2, (smaller_size - new_img_size[1])//2)) #pastes the image on black image in middle
        base=os.path.basename(list_img[img])
        path = os.path.splitext(base)[0]
        black_image.save(f'smaller_images/{path}.jpg')
        #black_image.save('{}{}'.format(fn, fext)) #saves in same directory
    # # black_image.save(f'image_{n}.jpg')
```
